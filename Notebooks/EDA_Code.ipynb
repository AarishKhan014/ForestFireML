```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(rf'D:\DS_Work\Projects\Regression\Algerian_forest_fires_CLEANED.csv')
# df.drop(columns=['day', 'month', 'year'], inplace=True)
# df['Classes'] = np.where(df['Classes'] == 'fire', 1, 0)
# df['Region'] = np.where(df['Region'] == 'Bejaia', 0, 1)
df.head()
```

```python
# Histplots

df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Pie Chart Of Classes

percentage = round(df['Classes'].value_counts(normalize=True)*100, 2)

classlabels = ['Fire', 'Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels, autopct='%1.2f%%')
plt.title ('Pie Chart Of Classes')
plt.show()
```

```python
# Correlation
plt.figure(figsize=(12,8))
sns.heatmap(df[df.columns[3:-2]].corr(), annot=True)
plt.show()
```

```python
# Monthly Fire Analysis of Sidi Belabbes & Bejaia

df_sidi = df[df['Region'] == 'Sidi Belabbes']
plt.subplots(figsize=(7,3.5))
sns.countplot(x='month', hue='Classes', data=df_sidi)
plt.xlabel('Months', weight='bold')
plt.ylabel('Number Of Fires & No_Fires', weight='bold')
plt.title ('Fire Data Of Sidi Belabbes', weight='bold')

df_bejaia = df[df['Region'] == 'Bejaia']
plt.subplots(figsize=(7,3.5))
sns.countplot(x='month', hue='Classes', data=df_bejaia)
plt.xlabel('Months', weight='bold')
plt.ylabel('Number Of Fires & No_Fires', weight='bold')
plt.title ('Fire Data Of Bejaia', weight='bold')
plt.show()
```

So Now The Data Is Concluded That The Rate of Fire Keeps Increasing From June to August and peaked in August and less In September.
